```python
c1 = pd.read_csv("/kaggle/input/Biggest%20Cities%20in%20the%20World.csv")
c2 = pd.read_csv("/kaggle/input/New_York_Tourist_Locations.csv")
```

```python
columns = "id city location lat lng parent population".split(" ")
c1 = c1.rename(columns = {"city_ascii": "location", "country": "parent"})
c1 = c1.drop_duplicates(subset = ['location'], keep = 'last')
```

```python
set_c = set()
for i in c1['parent']:
    set_c.add(i)
```

```python
new_df = pd.DataFrame([[0,i,i,0,0,i,0,] for i in set_c], columns = columns)
```

```python
c1 = pd.concat([c1, new_df])
c1['id'] = [i for i in range(len(c1['id']))]
c1
```

```python
c1
```

```python
c1 = c1.drop(columns = ['city','lng','lat'])
c1.reset_index(inplace = True)
c1
```

```python
c1 = c1.drop(columns = ['index'])
c1
```

```python
def get_id(string, df):
    return int(df['id'][df.index[df.location == string]].iloc[0])
def get_parent(string, df):
    return df['parent'][df.index[df.location == string]].iloc[0]
def get_parent_id(string, df):
    v = get_parent(string, df)
    if v == string:
        return -1
    else:
        return get_id(v, df)
```

```python
c1['parent_id'] = [get_parent_id(i, c1) for i in c1['location']]
c1
```

```python
c1 = c1.drop(columns = ['population'])
```
